```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
dados = pd.read_csv("/content/A102039189_28_143_208.csv", encoding="ISO-8859-1", 
                    skiprows = 3,
                    sep=";",
                    skipfooter=12,
                    thousands=".",
                    decimal=",",
                    engine='python')
```

```python
ax = dados.plot(x = 'Unidade da Federação', 
                y = '2021/Mar',
                kind = "bar",
                figsize = (14,7))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x/1000000) + '')) 

plt.title('Agendamentos Hospitalares do SUS por Unidade da Federação',
          fontsize = 18)
plt.xlabel('')
plt.ylabel('Procedimentos Hospitalares (milhões)',
           fontsize = 14)
plt.grid(axis = 'y', color = 'green', linestyle = '--', linewidth = 0.5)
```
